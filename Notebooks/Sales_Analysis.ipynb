```python
#### Q.2) Most Selling Product ?
```

```python
data.head()                  #- By Quantity
```

```python
data.groupby('Product')['Quantity'].sum() 
```

```python
data.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
```

```python
most_quantity = data.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
most_quantity
```

```python
type(most_quantity)
```

```python
most_quantity = most_quantity.reset_index() 
most_quantity
```

```python
type(most_quantity)
```

```python
plt.figure(figsize = (9,4))
plt.bar(most_quantity['Product'], most_quantity['Quantity'], color = ['red', 'black', 'green', 'yellow', 'cyan'], width=0.4)
plt.title("Most Selling Product - By Quantity")              
plt.xlabel("Product")
plt.ylabel("Quantity");
```

```python
data.head()          # By Revenue
```

```python
data['Revenue'] = data['Price'] * data['Quantity']
```

```python
data
```

```python
data.groupby('Product')['Revenue'].sum().sort_values(ascending = False)
```

```python
most_revenue = data.groupby('Product')['Revenue'].sum().sort_values(ascending = False)
most_revenue
```

```python
most_revenue = data.groupby('Product')['Revenue'].sum().sort_values(ascending = False)
most_revenue
```

```python
plt.bar(most_revenue.index, most_revenue.values,color=['green', 'red', 'black', 'yellow', 'cyan'], width=0.3)

plt.title("Most Selling Product - By Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()
```
